### Objectif: Il s’agit de prédire le prix de biens immobiliers en fonction d’informations concernant l’environnement en utilisant le dataset BOSTON. On va Predire le MEDV (- Median value of owner-occupied homes in 1000's dollar ) selon les autres variables

```r
autos=read.csv("boston.csv")
autos
```

```r
summary(autos)
```

```r
Modelreg <- lm( MEDV ~ CRIM + ZN + INDUS+ CHAS + NOX + RM + AGE + DIS + RAD + TAX +PTRATIO +LSTAT, autos)
```

```r
Modelreg
```

```r
summary(Modelreg)
```

=> les résidus sont distribués selon une loi normale et
les valeurs |Q1|=0.4758 et |Q3|=0.4845 sont tres proches

```r
plot(Modelreg)
```

=> la premiere plot nous montre que lorsque les réponses prédites par le modèle augmentent les résidus restent globalement uniformément distribués de part et d’autre de
0. Cela nous montre que la droite de régression est bien adaptée aux données alors notre hypothèse de linéarité est acceptée.

=> la 2eme plot: les résidus sont bien distribués le long de la droite figurant sur le plot ; l’hypothèse de
normalité est acceptée.

=> la 3eme plot: la courbe en rouge est globalement ascendante ou descendante, alors  la
dispersion des résidus n’est pas homogène.dans ce cas l’hypothèse est rejetée.

```r
install.packages('MASS')
library(MASS)
```

```r
#Modèle Réduit
modeleRéduit<- stepAIC (Modelreg , scope = list (lower =" MEDV ~1 " , upper = "MEDV ~
CRIM + ZN + INDUS+ CHAS + NOX + RM + AGE + DIS + RAD + TAX +PTRATIO +LSTAT" ) , direction = "backward" )
```
